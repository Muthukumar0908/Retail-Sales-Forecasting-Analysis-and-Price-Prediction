# Odinial encoding 

```python
from sklearn.preprocessing import OrdinalEncoder
enc=OrdinalEncoder()
for i in df1.columns:
    if df1[i].dtype == 'object':
#         col_name = i
        df1[i] = enc.fit_transform(df1[[i]])
        encoding = df1[i].sort_values().unique()
        globals()[i] = {}
        decoded_data = enc.inverse_transform(df1[[i]])
        decoded = sorted(np.unique(decoded_data)) 
        globals()[i] = dict(zip(decoded, encoding))
```

```python
for i in df1.columns:
    if df1[i].dtype == 'bool':
#         col_name = i
        df1[i] = enc.fit_transform(df1[[i]])
        encoding = df1[i].sort_values().unique()
        globals()[i] = {}
        decoded_data = enc.inverse_transform(df1[[i]])
        decoded = sorted(np.unique(decoded_data)) 
        globals()[i] = dict(zip(decoded, encoding))
```

# Random forest

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=90,random_state=15,max_depth=21,min_samples_split=10)
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mean_squared_error(y_pred,y_test)
```

```python
mean_absolute_error(y_pred,y_test)
```
