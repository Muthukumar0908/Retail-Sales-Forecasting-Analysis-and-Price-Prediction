##  Hypothesis testing

# catagri vs catagri more then two columns 

```python
import numpy as np
from scipy.stats import chi2_contingency
for i in data.columns:
    for j in data.columns:
        table=pd.crosstab(df[i],df[j])
        chi2, p,dof, expected = chi2_contingency(table)
        if p >0.05:
            print(i,j,p)
```
